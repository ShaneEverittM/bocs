# Analyse the output

```python
OUTPUT_DIR = "../../EdgePart2Examples"

class OutputComparator:
    def __init__(self, cms_file, prof_file):
        self.cms_file = cms_file
        self.prof_file = prof_file
        _read()
        
    def _read(self):
        with open(path.join(OUTPUT_DIR, self.cms_file), "r") as cf:
            self.cms_output = cf.readlines()

        with open(path.join(OUTPUT_DIR, self.prof_file), "r") as pf:
            self.actual_output = pf.readlines()
        
        
```

```python
OUTPUT_DIR = "../../EdgePart2Examples"

with open(path.join(OUTPUT_DIR, "blant-cms.k7"), "r") as cms_file:
    cms_output = cms_file.readlines()

with open(path.join(OUTPUT_DIR, "blant-sum.k7"), "r") as prof_file:
    actual_output = prof_file.readlines()
```

```python
for i, line in enumerate(actual_output):
    actual_output[i] = line.strip().split("\t")
    
for i, line in enumerate(cms_output):
    cms_output[i] = line.strip().split("\t")
```

```python
actual_output = sorted(actual_output, key=lambda a: a[0])
cms_output = sorted(cms_output, key=lambda a: a[0])
```

```python
for i, line in enumerate(actual_output):
    actual_output[i][1:] = sorted(line[1:])

for i, line in enumerate(cms_output):
    cms_output[i][1:] = sorted(line[1:])
```

```python
assert(len(cms_output) == len(actual_output))
```

```python
mismatched_lines = 0
mismatched_cells = 0
for i, cms_line in enumerate(cms_output):
    actual_line = actual_output[i]
    if len(cms_line) != len(actual_line):
        print(f"Line mismatch {i}: {cms_line} {actual_line}")
        mismatched_lines += 1
        continue
    for j, cms_cell in enumerate(cms_line):
        actual_cell = actual_line[j]
        if actual_cell != cms_cell:
            mismatched_cells += 1
            print(f"Cell mismatch {i}: {cms_cell} {actual_cell}")

print(mismatched_lines, mismatched_cells)
```
